```python
from nltk.corpus import stopwords
def no_user_alpha(tweet):
    clean_mess = [word for word in tweet.split() if word.lower() not in stopwords.words('english')]
    return clean_mess

print(no_user_alpha(new_df['text'].iloc[10]))
```
